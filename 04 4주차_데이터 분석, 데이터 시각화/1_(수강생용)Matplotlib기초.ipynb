```python
plt.plot(df.norm1)
plt.show()
```

```python
# 그래프 축 레이블 설정
# labelpad : 그래프와 레이블 사이의 여백 크기
# fontdict : 폰트 타입과 스타일 관련 설정 
# fontdict의 키워드 : color, weight(bold, normal), size

data_x = np.array([1, 2, 3, 4])
data_y = data_x **3

plt.plot(data_x, data_y)
plt.xlabel('X Axis Label', labelpad=12, fontdict={'color': 'b', 'size': 12})    #  폰트타입 설정 >> 'family' : 'arial'
plt.ylabel('Y Axis Label', labelpad=12, fontdict={'color': 'k', 'size': 8})          #  폰트두께 설정 >> 'weight' : 'normal'  
plt.show()
```

```python
# 범례 표시하기
# plot 함수에서 label 키워드 : 범례 레이블
# legend 함수 : 범례 레이블 표시 
# legend 함수의 loc 키워드 : 위치 지정 ex) 'best', (1.0, 1.0) 으로 설정 가능

plt.plot(data_x, data_y, label='x^3')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend()       # 위치 숫자로 입력 가능 legend(loc=(1.0, 1.0))
plt.show()
```

```python
# 범례를 여러 개 설정 가능함.
# ncol=2 ; 2개 범례 표시

plt.plot(data_x, data_y, label='x^3')
plt.plot(data_y, data_x, label='cube root x')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc="best", ncol=2)  
plt.show()
```

> loc 입력 가능 옵션

- "best"
- "upper right"
- "upper left"
- "lower left"
- "lower right"
- "right"
- "center left"
- "center right"
- "lower center"
- "upper center"
- "center"

```python
# x축, y축 범위 설정
# xlim, ylim 함수 활용

plt.plot(data_x, data_y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.xlim([0, 5])   
plt.ylim([-2, 40])    
plt.show()
```

```python
# axis 함수를 이용한 범위 설정
# plt.axis([x축 최소값, x축 최대값, y축 최소값, y축 최대값])

plt.plot(data_x, data_y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.axis([0, 5, -2, 40]) 
plt.show()
```

```python
# matplotlib 기본 컬러 팔레트 10가지 색상을 이용한 출력
# 참고사이트: https://matplotlib.org/stable/gallery/color/named_colors.html

plt.plot(data_x, data_y, color='r')    
plt.show()
```

```python
# 라인 스타일 설정하기
# plot 함수에서 linestyle 키워드를 사용함.
# 기본 설정은 solid 타입임.
# 'solid', 'dashed', 'dotted', 'dashdot' 등을 value로 사용할 수 있음.
# 참고자료: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html


x = np.array(range(1,10))

plt.plot(x, 2*x, linestyle='solid', color='r', label='solid')
plt.plot(x, 3*x, linestyle='dashed', color='g', label='dashed')
plt.plot(x, x**2, linestyle='dotted', color='b', label='dotted')
plt.plot(x, x**3, linestyle='dashdot', color='k', label='dashdot')

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='best')

plt.axis([0.5, 4.5, 0, 20])
plt.show()
```

```python
# linewidth: 라인 두께 설정하기
# for linestyle='solid' -> sold_capstyle 지정
# for linestyle='dash' or 'dotted' and some like -> dash_capstyle 지정

plt.plot(x, x**2, linestyle='dotted', color='g', label='dotted', dash_capstyle='round', linewidth=8)
plt.plot(x, x**3, linestyle='dashdot', color='k', label='dashdot', dash_capstyle='butt', linewidth=8)
plt.plot(x, x**4, color='k', label='solid', solid_capstyle='round', linewidth=8)

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='best')

plt.axis([0.5, 4.5, 0, 9])
plt.show()
```

```python
# marker 지정하기1

plt.plot(x, x**2, linestyle='dotted', marker='o', label='dotted')
plt.plot(x, x**3, linestyle='dashdot', marker='^', color='k', label='dashdot')

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='best')

plt.axis([0.5, 4.5, 0, 9])
plt.show()
```

```python
# marker 지정하기2
# 이띠 linestyle은 지정해 주지 않는다는 것.

plt.plot(x, x**2, 'o--', label='solid_marker')
plt.plot(x, x**3, 'x:', color='k', label='dashed_marker')

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='best')

plt.axis([0.5, 4.5, 0, 9])
plt.show()
```

```python
# 격자 그리기

plt.plot(x, x**2, 'o--', label='solid_marker')
plt.plot(x, x**3, 'x:', color='k', label='dashed_marker')

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='best')
plt.grid()

plt.axis([0.5, 4.5, 0, 9])
plt.show()
```

```python
# 그리드 설정
# axis: 축지정
# color: 그리드 컬러
# alpha: 그리드 투명도
# linestyle: 그리드 라인스타일

plt.plot(x, x**2, 'o--', label='solid_marker')
plt.plot(x, x**3, 'x:', color='k', label='dashed_marker')

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='best')
plt.grid(True, axis='y', color='orange', alpha=0.25, linestyle='-.')

plt.axis([0.5, 4.5, 0, 9])
plt.show()


```

```python
# 타이틀 설정하기
# title 함수를 이용한 그래프 제목 설정
# title의 위치(loc), 그래프와의 간격(pad), 폰트스타일(fontdict) 등을 설정할 수 있음.
# 폰트스타일은 키워드 fontdict를 이용하여 한꺼번에 설정해도 됨. 'fontsize', 'fontweight' 

plt.plot(x, x**2, 'o--', color='g', label='solid_marker')

plt.legend(loc='best')
plt.grid(axis='y', alpha=0.25)

font_style = {
    'fontsize': 14,
    'fontweight': 'bold'   #  {‘normal’,‘bold’, 'heavy’,‘light’,‘ultrabold’,‘ultralight’} 설정 가능
}

plt.title('Title of Chart', fontdict=font_style, loc='center', pad=15)
plt.show()
```

```python
# 여러 그래프 동시에 그리기
# subplot() 함수 활용
# 먼저 그리고자 하는 레이아웃을 몇 개의 행 또는 열로 나눌 것인지 기준을 정해야 함.
# subplot(행, 열, 인덱스)


plt.subplot(2, 1, 1)
plt.plot(x, x**2, linestyle='solid', color='g', label='solid')
plt.xlabel('x')
plt.ylabel('x ^ 2')
plt.xlim((0, 10))
plt.ylim((0, 800))

plt.subplot(2, 1, 2)
plt.plot(x, x**3, linestyle='solid', color='limegreen', label='solid')
plt.xlabel('x')
plt.ylabel('x ^ 3')
plt.xlim((0, 10))
plt.ylim((0, 800))

plt.tight_layout()
plt.show()
```

```python
ax1 = plt.subplot(1, 2, 1)
ax1.plot(x, x**2, linestyle='solid', color='g', label='solid')
ax1.set_xlabel('x')
ax1.set_ylabel('x ^ 2')
ax1.set_xlim((0, 10))
ax1.set_ylim((0, 800))

ax2 = plt.subplot(1, 2, 2)
ax2.plot(x, x**3, linestyle='solid', color='limegreen', label='solid')
ax2.set_xlabel('x')
ax2.set_ylabel('x ^ 3')
ax2.set_xlim((0, 10))
ax2.set_ylim((0, 800))

plt.tight_layout()
plt.show()
```

# 2.hist()

```python
# seaborn의 iris dataset 활용

iris = sns.load_dataset('iris')
iris.head()
```

```python
# 데이터 분포 확인을 위한 히스토그램 그리기
# hist() 함수 사용

plt.hist(iris.sepal_length, color='skyblue')
plt.show()
```

```python
# 두 가지 이상의 변수 히스토그램 그리기
# plot.hist() 함수를 변수를 달리하면 연속적으로 사용
# 마지막으로 그림을 표현할 때, plt.show() 함수 사용

plt.hist(iris.sepal_length, bins=20, label='sepal_length', color='skyblue')
plt.hist(iris.petal_length, bins=20, label='petal_length', color='violet')
plt.legend(loc='upper right')
plt.show()
```

```python
# 누적 히스토그램 그리기
# 동일하게 plt.hist() 함수를 사용하는데, cumulative=True로 설정하면 됨.

plt.hist(iris.sepal_length, cumulative=True, label='cumulative', color='skyblue')
plt.hist(iris.sepal_length, cumulative=False, label='not cumulative', color='violet')
plt.legend(loc='upper left')
plt.show()
```

```python
# 히스토그램의 4가지 타입
# plt.hist() 함수의 파라미터 histtype에 'bar', 'barstacked', 'step', 'stepfilled' 4가지를 설정

fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3) 
ax4 = fig.add_subplot(2, 2, 4)

ax1.hist((iris.sepal_width, iris.petal_width), histtype='bar')
ax2.hist((iris.sepal_width, iris.petal_width), histtype='barstacked')
ax3.hist((iris.sepal_width, iris.petal_width), histtype='step')
ax4.hist((iris.sepal_width, iris.petal_width), histtype='stepfilled')

ax1.title.set_text('Type - bar')
ax2.title.set_text('Type - barstacked')
ax3.title.set_text('Type - step')
ax4.title.set_text('Type - stepfilled')

plt.tight_layout()
plt.show()
```

# 3.bar()

```python
# 범주형 데이터 개수 확인을 위한 막대그래프
# 범주형 데이터를 만들기 위해, 'species'로 그룹을 묶고, petal_length에 대해 평균을 구한 값

iris_petal_length_mean = iris.groupby('species').petal_length.mean()
iris_petal_length_mean
```

```python
# iris species에 대한 label 값

iris.species.unique()
```

```python
# bar 그래프는 x축에 범주형 데이터 label, y축에는 범주형 데이터 label에 대한 수치값이 적용됨.

plt.bar(iris.species.unique(), iris.groupby('species').petal_length.mean())
plt.show()
```

```python
plt.bar(iris.species.unique(), iris_petal_length_mean, color='mediumseagreen')
plt.show()
```

```python
# 범주형 데이터 label 마다 color를 개별적으로 적용할 수 있음.
# 파라미터 color에 label의 수많큼 컬러를 리스트에 담아 전달
# 파라미터 width를 막대 너비 조절. 최대 max가 1임.

plt.bar(iris.species.unique(), iris_petal_length_mean, color=['gold', 'mediumseagreen', 'teal'],
       width=0.98)
plt.show()
```

```python
# 수평 형태의 bar를 그리기 위해 barh() 함수를 적용
# 매개변수도 bar()와 동일하게 사용함.

plt.barh(iris.species.unique(), iris_petal_length_mean)
plt.show()
```

# 4.scatter()

```python
# 수치형 데이터의 상관관계를 파악할 수 있는 산점도
# x, y축 각각에 변수를 설정하고 두 개의 값에 대응하는 위치를 원으로 표현함.
# 두 변수 간의 관계가 선형인지 비선형인지 확인하여 방향성과 강도를 파악할 수 있음.
# 파라미터 x, y는 생각하고 순서대로, x축에 사용할 특성, y축에 사용할 특성을 적음.

plt.scatter(x=iris.petal_length, y=iris.petal_width)
plt.show()
```

```python
# scatter 도형의 사이즈, 컬러, 투명도는 각각 s, c, alpha로 변경할 수 있음.
# cmap은 color 맵을 의미함.

area = (18 * np.random.rand(len(iris.petal_length)))**2
colors = np.random.rand(len(iris.petal_length))

plt.scatter(iris.petal_length, iris.petal_width, s=area, c=colors, alpha=0.75, cmap='Set1_r')
plt.show()
```

```python
# 값의 위치에 따라 표현되는 컬러 팔레트를 이용한 color 설정하기
# 컬러맵 참고자료: https://matplotlib.org/stable/tutorials/colors/colormaps.html

import matplotlib.cm as cm
colors = cm.RdYlGn(np.linspace(0, 1, iris.petal_width.shape[0]))

plt.scatter(iris.petal_length, iris.petal_width, s=10**2, c=colors)
plt.show()
```

```python
# 데이터 그룹별로 다른 컬러를 적용하는 경우

# 먼저, iris 세 종류를 개별 데이터로 만듦.
setosa = iris[iris.species == 'setosa']
versicolor = iris[iris.species == 'versicolor']
virginica = iris[iris.species == 'virginica']

# plt.scatter() 함수를 이용하여 개별 데이터를 시각화함. 동일한 셀에서 3개의 scatter가 실행되어야 함.
plt.scatter(setosa.sepal_length, setosa.petal_width, color = 'hotpink')
plt.scatter(versicolor.sepal_length, versicolor.petal_width, color = '#88c999')
plt.scatter(virginica.sepal_length, virginica.petal_width, color = 'skyblue')

plt.show()
```

# 5.boxplot()

```python
# 데이터 분포 확인과 이상값 찾기를 위한 박스플랏
# 보통 데이터 프레임의 각 열에 대해 플랏을 수행하며, 수치형 데이터 타입을 대상으로 함.

plt.boxplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width])
plt.show()
```

```python
# 매개변수 whis는 기본값이 1.5인데, 변경 가능함. 
# whis는 이상치 처리 기준 설정 시, IQR의 배수를 의미함.
# notch를 True로 입력하면 중앙값의 95% 신뢰구간을 노치 형태로 표시한다는 의미임.

plt.boxplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width], 
            notch=True, whis=2.0)
plt.show()
```

```python
# boxplot()은 그래프 출력 외에도 해당 변수의 분위수와 중앙값을 반환함.
# whisker는 min, QR1, QR3, max를 반환
# median은 중앙값
# fliers는 수염을 벗어나는 포인트인 이상값을 반환함.
# plt.boxplot()을 객체로 만든 다음, key 형태로 결과를 받아옴.

boxplot = plt.boxplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width])

whiskers = [item.get_ydata() for item in boxplot['whiskers']]
medians = [item.get_ydata() for item in boxplot['medians']]
fliers = [item.get_ydata() for item in boxplot['fliers']]

plt.show()
```

```python
print('whiskers:', whiskers)
```

```python
print('medians:', medians)
```

```python
print('fliers:', fliers)
```

```python
# 수평 형태의 박스플롯 그리기
# 파라미터 vert=False로 설정하면 수평 형태 박스플롯을 그림

plt.boxplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width], vert=False)
plt.show()
```

# 6.violinplot()

```python
# 데이터 분포 모양과 상세 모양 확인을 위한 바이올린그래프 그리기
# plt.violinplot() 함수 사용
# boxplot과 사용법이 비슷함.
# 파라미터 positions는 그래프 출력 순서를 의미함. default는 [1,2,3,4]

plt.violinplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width], positions=[2, 1, 4, 3])
plt.show()
```

```python
# 파라미터 showmeans=True로 설정하면, 중앙값을 표시함

plt.violinplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width], showmeans=True)
plt.show()
```

```python
# violinplot() 함수는 데이터 분포 분위수를 직접 리스트로 입력해서 표시할 수 있음.
# 파라미터 quantiles에서 중첩 리스트 형태로 전달됨.
# 아래 예시는 sepal_length violinplot에 대해 25%, 75%에 표시하는 것임.

plt.violinplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width], 
               quantiles=[[0.25, 0.75], [0.1, 0.9], [0.2, 0.8], [0.35, 0.65]])
plt.show()
```

```python
# violinplot 컬러 변경하기
# violinplot() 함수는 color 파라미터가 없기 때문에 다른 방법으로 접근해야 함.

# violinplot() 객체를 만들고, bodies에 면적 색상을 적용함.
violinplot = plt.violinplot([iris.sepal_length, iris.sepal_width, iris.petal_length, iris.petal_width], showmeans=True)

violinplot['bodies'][0].set_facecolor('hotpink')
violinplot['bodies'][1].set_facecolor('purple')
violinplot['bodies'][2].set_facecolor('limegreen')
violinplot['bodies'][3].set_facecolor('darkgreen')

# 플롯의 라인 컬러 변경. 순서: 세로 라인, 최댓값, 최솟값, 중앙값 표시 라인
violinplot['cbars'].set_edgecolor('k')
violinplot['cmaxes'].set_edgecolor('darkgrey')
violinplot['cmins'].set_edgecolor('darkgrey')
violinplot['cmeans'].set_edgecolor('k')

plt.show()
```
