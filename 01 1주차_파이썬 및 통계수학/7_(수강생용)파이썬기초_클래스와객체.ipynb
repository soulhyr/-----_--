# 1. 파이썬의 클래스


## 1.1 객체 지향 프로그래밍

> 파이썬은 객체지향프로그래밍(OOP) 언어로서, 객체지향프로그래밍의 주요 원칙과 개념을 지원하고 사용함. <br>
이를 통해 코드를 구조화하고 관리하기 쉽게 만들어줌.

- 객체(Object)
    - 파이썬은 모든 것을 객체로 다룸.
    - 숫자, 문자열, 리스트, 함수 등 모든 것이 객체로 취급됨.
    - 객체는 속성(attribute)와 메서드(method)를 가질 수 있으며, 클래스에 의해 정의됨.
- 클래스(Class)
    - 클래스는 객체를 생성하기 위한 템플릿 또는 설계도 역할을 함.
    - 클래스는 속성과 해당 속성을 처리하기 위한 메서드를 정의함.
    - 파이썬에서 클래스는 class 키워드를 사용하여 정의됨.
- 인스턴스(Instance)
    - 클래스를 기반으로 생성된 실제 객체를 인스턴스라고 함.
    - 각 인스턴스는 고유한 속성과 메서드를 가질 수 있음.
    - 예를 들어, int 클래스의 인스턴스는 정수 값을 나타내며, str 클래스의 인스턴스는 문자열을 나타냄.
- 캡슐화(Encapsulation)
    - OOP의 중요한 개념 중 하나는 캡슐화임.
    - 파이썬은 클래스 내에서 속성과 메서드를 하나로 묶어 캡슐화를 지원함.
    - 이로써 속성과 메서드의 응집성을 높이고 외부에서 직접 접근하지 않도록 보호할 수 있음.
- 상속(Inheritance)
    - 파이썬은 상속을 통해 새로운 클래스가 기존 클래스의 속성과 메서드를 상속받을 수 있도록 지원함.
    - 이는 코드 재사용을 촉진하고 계층 구조를 형성하는 데 도움이 됨.

## 1.2. 파이썬 클래스와 객체

- 클래스가 필요한 이유
  - 여러 개의 객체를 동시에 만들어서 사용할 필요가 있을 때, 함수보다 더 효율적임.
  - 더군다나 그 객체마다 특성, 메소드를 구비하고 있어 관련 데이터를 저장하고 액션을 취하기에 편함.
  - 그러나 클래스가 없어도 프로그램을 작성에 문제는 없음. 대표적으로 C 언어는 클래스 개념이 없음.

- 클래스와 객체
  - 클래스는 붕어빵 틀, 객체는 붕어빵 틀로 찍어낸 붕어빵이라 할 수 있음.
  - 클래스(class)란 똑같은 무언가를 만들어 낼 수 있는 설계 도면과 같음.
  - 객체(object)란 클래스로 만들어진 결과물과 같음.
  - 클래스로 만든 객체는 객체마다 고유한 성격을 가지고 독립적임.

## 1.4. 클래스 생성자(\_\_init__ method) 구현하기
1. 클래스는 무작정 만드는 것이 아니라, 구조와 기능을 생각하고 먼저 설계해야 함.
2. 클래스 내부에는 함수 기능을 하는 메소드(method), 변수(상수 포함) 기능을 하는 특성(attribute)으로 구성됨.
3. 생성자(constructor)란 객체가 생성될 때 자동으로 호출되는 메서드를 의미함.
 - 파이썬 메서드명으로 \_\_init__을 사용함.
 - self는 클래스 객체 자신을 의미함. class의 모든 메소드에 self는 기본 파라미터로 들어감.
4. 생성자는 클래스 생성 시 필수적인 기본 환경 세팅 과정이라고 생각하면 됨.


````python
# 클래스 생성자 만들기
# self가 첫번째 매개변수로 들어간다는 점 유의

class ClassRoom:
    def __init__(self):
        self.print = "welcome to my classroom"
````

````python
class1 = ClassRoom()

class1.print
````

## 1.5. 클래스 메서드 \_\_call()__ 구현하기
1. 클래스 객체를 함수처럼 사용할 수 있는 메서드
2. 클래스 메서드와는 다르게 객체 자체를 함수로 사용할 수 있음.


````python
# 클래스 생성자 만들기
# self가 첫번째 매개변수로 들어간다는 점 유의

class ClassRoom:
    def __init__(self, num, field):
        self.class_num = num
        self.field = field

    def desc(self):
        print(f'우리반은 {self.field}를 배우는 반입니다.')

    def __call__(self):
        print(f'우리반에 오신것을 진심으로 환영합니다.')
````

````python
# 객체 함수를 실행하면, 메서드 \_\_call__ 이 실행됨.

nlp_class = ClassRoom(7, 'nlp')

# 일반 메서드와 __call()__의 차이
nlp_class.desc()
nlp_class()
````

````python
# __call__() 파라미터 설정하기

class ClassRoom:
    def __init__(self, num, field):
        self.class_num = num
        self.field = field

    def desc(self):
        print(f'우리반은 {self.field}를 배우는 반입니다.')

    def __call__(self, mood):
        print(f'우리반에 오신것을 진심으로 {mood} 환영합니다.')
````

````python
nlp_class = ClassRoom(7, 'nlp')

nlp_class.desc()
nlp_class('열열히')
````

## 1.6. 클래스 메서드 구현하기
1. 클래스 내부에는 객체마다 액션을 위한 함수를 설계할 수 있는데, 이를 메서드라고 함.
2. 클래스 또는 객체 내에서 실행 가능한 기능 함수라고 할 수 있음.
3. 메서드의 첫번째 파라미터 또한 self 임.

````python
# 클래스 생성자 만들기
# 클래스 메서드 만들기

class ClassRoom:
    def __init__(self, num, field):
        self.class_num = num
        self.field = field

    def print_num(self):
        print(f'우리반은 {self.class_num}반입니다.')
        return self.class_num

    def print_field(self, mood):
        print(f'우리반은 {self.field} 분야를 {mood} 배웁니다.')
````

## 1.7. 클래스 상속
1. 상속(Inheritance)이란 ‘물려받다’라는 뜻으로, 클래스를 만들 때, 다른 클래스의 기능을 '물려받아' 생성함을 의미함.
2. 조금씩 다른 여러 클래스를 만들 때, 공통된 기능을 모아 basic class를 구축해 놓으면, 상속을 통해 쉽고 빠르게 다양한 클래스를 설계할 수 있음.
3. 보통 상속은 기존 클래스를 변경하지 않고 기능을 추가하거나 기존 기능을 변경하려고 할 때 사용함.
4. 쉽게는 기존 클래스를 수정하면 되지만 기존 클래스가 라이브러리 형태로 제공되거나 수정이 허용되지 않는 상황이라면 상속을 사용해야 함.
3. 구조:
```python
 class 클래스_이름(상속할_클래스_이름)
```

````python
# basic class를 ClassRoom으로 설계함.
# ClassRoom을 그대로 상속하는 NNPClass 설정

class ClassRoom:
    def __init__(self, num, field):
        self.class_num = num
        self.field = field

    def print_num(self):
        print(f'우리반은 {self.class_num}반입니다.')
        return self.class_num

    def print_field(self, mood):
        print(f'우리반은 {self.field} 분야를 {mood} 배웁니다.')


class NLPClass(ClassRoom):
    pass


class NLPClass_dummy():
    pass
````

````python
# 상속 받은 NLPClass 예시

nlp_class7 = NLPClass(10, 'nlp')

print(nlp_class7)
print(nlp_class7.class_num)
print(nlp_class7.print_num())
print(nlp_class7.print_field('꿈을 품고'))
````

````python
# 상속 받지 않은 NLPClass_dummy

nlp_class_dummy = NLPClass_dummy(10, 'nlp')

print(nlp_class_dummy)
````

### 1.7.1. 메서드 오버라이딩
1. 부모 클래스(상속한 클래스)에 있는 메서드를 동일한 이름으로 다시 만드는 것을 메서드 오버라이딩(method overriding)이라고 함.
2. 이렇게 메서드를 오버라이딩하면 부모 클래스의 메서드 대신 오버라이딩한 자식 메서드가 우선 호출됨.
3. 만약 메서드 오버라이딩을 활용하지 않으면 기존 부모 클래스의 메서드를 그대로 활용할 수 있음.

````python
# basic class를 ClassRoom으로 설계함.
# ClassRoom을 그대로 상속하는 NLPClass_overriding 설정
# __init__() overriding. field를 입력받지 않아도 'NLP' 자동 설정되도록 overring 함.
# 메서드 pring_num() overriding.
# 메서드 print_filed()는 부모 클래스 메서도 그대로 이용함.

class ClassRoom:
    def __init__(self, num, field):
        self.class_num = num
        self.field = field

    def print_num(self):
        print(f'우리반은 {self.class_num}반입니다.')

    def print_field(self, mood):
        print(f'우리반은 {self.field} 분야를 {mood} 배웁니다.')


class NLPClass_overriding(ClassRoom):
    def __init__(self, num):
        self.class_num = num
        self.field = 'NLP'

    def print_num(self):
        print(f'우리반은 {self.field} {self.class_num}반입니다.')
        print(f'메서드 오버라이드, 어렵지 않쥬??')


class VisionClass_overriding(ClassRoom):
    pass
````

````python
# ClassRoom 객체 생성 및 예시

vision10 = ClassRoom(10, 'vision')

vision10.print_num()
````

````python
# NLPClass_overriding 객체 생성 및 예시
# overriding 한 메서드를 호출하였기 때문에 자식 클래스의 메서드가 실행됨.

nlp8_overriding = NLPClass_overriding(8)

nlp8_overriding.print_num()
````

````python
# overriding 하지 않은 메서드를 호출하면, 부모 클래스의 메서드가 실행됨.
# object.print_field()

nlp8_overriding.print_num() # overring 적용
print('-'*30)
nlp8_overriding.print_field('꿈을 품고') # 부모 클래스 메서드를 그대로 활용함.
````

````python
# 클래스 생성

class YourName():
    lastname = 'Ha'
````

````python
ha = YourName()
kim = YourName()

print(ha.lastname)
print(kim.lastname)
````

### 1.7.3. 부모 클래스 메서드 호출하기
- super() 활용
- 자식 클래스가 부모 클래스의 특정 메서드를 오버라이딩하면, 자연스레 자식 클래스에서는 부모 클래스의 동일 메서드를 쓰지 못함.
- 그러나 super()를 이용하여 부모 클래스의 특정 메서드를 호출하여 사용할 수 있음.

````python
class ParentClass():
    def hobby(self):
        print('reading a book')

class ChildClass(ParentClass):
    def hobby(self):
        print('playing soccer')
        print('my parent hobby is, ')
        super().hobby()

boy = ChildClass()
boy.hobby()
````
