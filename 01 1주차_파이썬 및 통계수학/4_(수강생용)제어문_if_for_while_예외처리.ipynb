# **제어문(조건문과 반복문)**
----

> 제어 구조(control structure)
* 프로그래밍에서 제어란 데이터의 흐름을 바꾸는 것을 의미함.
* 제어 구조의 종류
  * 순차 구조
    - 데이터가 순서대로 처리되는 구조
  * 분기 구조
    - 데이터가 조건에 의해 흐름이 분리되는 구조
  * 반복 구조
    - 데이터의 흐름을 되돌리는 구조
  * 함수 구조
    - 특정 제어 구조를 묶어 모듈화 시키는 구조


> 조건문
* 분기하거나 반복하기 위한 조건을 따질 때 쓰는 **문**(statement)을 **조건문**(conditional statements) 또는 **if문** 이라고 함.
* 프로그램에서 특정한 조건을 검사하고 그 결과에 따라 프로그램의 실행 흐름을 변경할 수 있도록 하는 기능 <br>
* 조건문에는 if, if...else, if...elif...else문이 있음.
* 조건에 따라 어떤 코드를 실행할지 결정하는 세 가지 방법
    - if: 조건이 **True**이면 동작을 수행, **False**이면 무시하고 지나감.
    - if...else: 조건이 **True**이면 동작을 수행, **False**이면 else 아래 명령어를 수행함.
    - if...elif...else: 여러 조건들의 참, 거짓에 따라서 다르게 동작함.

> 반복문
* 동일한 작업을 여러 번 수행해야 할 때 사용하는 제어문
* 프로그램의 코드를 단순화하고 작업 효율성을 높일 수 있음.
* 조건문과 반복문이 복합된 while, for문은 반복문이라고 통칭함.
* 다른 프로그래밍 언어와 달리, 'break'와 'continue' 키워드를 통해 반복문의 제어가 가능함.
* 'break'는 반복문을 종료시키고, 'continue'는 현재 반복을 종료하고 다음 반복으로 넘어가도록 함.

## 1.3. if...else문과 if...elif...else문
> if...else문의 해석
* 만약 ~라면(if)
*   ~하고
* 그렇지 않다면(else)
*   ~한다.

> if...elif...else문의 해석
* 만약 ~라면(if)
*   ~하고
* 그렇지 않고 ~라면(elif)
*   ~하고
* 이 모든 조건에 맞지 않으면(else)
*   ~한다.

```python
# if...else문 활용 기초 예시1

score = 90

if score >=60:
    result = 'passed'
else:
    result = 'failed'
print(result)
```

```python
# 위와 완전히 동일한 코드

score = 90
result = ('passed' if score >=60 else 'failed') # 괄호가 반드시 필요한 것은 아니나 가독성이 좋아짐.
print(result)
```

```python
# if...else문 활용 기초 예시2

score = 50

if score >=60:
    result = 'passed'
else:
    result = 'failed'
    print('조금 더 힘을 내 봅시다 ^^')
print(result)
```

```python
# if...elif...else문 활용 예시1

num = 0

if num > 0:
    print("양수입니다")
elif num == 0:
    print("0입니다")
elif num == 3:
    print("3입니다")
else:
    print("음수입니다")
```

```python
# if...elif...else문 활용 예시2

num = int(input("숫자를 입력하세요: "))

if num > 0:
    print("양수입니다")
elif num == 0:
    print("0입니다")
else:
    print("음수입니다")
```

```python
# if...elif...else문 활용 예시3

score = int(input('점수를 입력해 주세요~^^'))

if score < 60:
    print('D')
elif score < 70:
    print("C")
elif score < 80:
    print("B")
elif score < 90:
    print("A")
else:
    print('S')
```

## 1.5. 중복 if문
* if 문이나 if...else 문을 겹쳐서 사용함.
* 이를 **중첩 if 문**(nested if statements)라고 함.
* 들여쓰기에 주의할 것.

```python
money = True
job = False
skill = None

if not money: # 돈이 없어? 라고 묻는 것과 동일함.
    print("make a money")
    if job: # 잡이 있어? 라고 묻는 것과 동일함.
        print('save a money')
    else:
        if skill:
            print('get a job')
        else:
            print('go to job skill school')
else:
    print('plan a trip')
```

# 2. for 반복문
* 반복 횟수가 미리 정해진 경우에 주로 사용
* 시퀀스 객체(문자열, 리스트, 튜플 등)의 각 요소를 하나씩 순회하면서 코드를 반복 실행하는 반복문

## 2.1. for 반복문 예시

```python
# iterable list

total = 0

for number in [2, -3, 0, 14, 10]:
    total = total + number
print(total)
```

```python
# iterable tuple

total = 0

for number in (2, -3, 0, 14, 10):
    total = total + number
print(total)
```

```python
# iterable string

string = "It is really hot!!"

for word in string:
    print(word, end='^^\n')
```

```python
# += 사용 예시
# -= 사용 예시

total = 0
for num in [1,2,3,4,5]:
    total += num
    print(total)

print('-'*50)

total = 0
for num in [1,2,3,4,5]:
    total = total + num
    print(total)
```

```python
# *= 사용 예시

total = 1

for num in [1,2,3,4,5]:
    total *= num
    print(total)
```

```python
# *= 사용 예시

total = 2

for num in [1,2,3,4,5]:
    total **= num
    print(total)
```

```python
# 리스트의 모든 요소 출력하기

fruits = ["apple", "banana", "cherry", 'melon', 'blueberry']

for fruit in fruits:
    print(fruit)
```

```python
# range() 함수를 사용하여 숫자 출력하기

for i in range(5):
    print(i)
```

```python
# range() 함수 사용하기
# range(데이터수)
# range(시작점, 최종점, step)

for i in range(1, 10, 2):
    print(i)
```

```python
# 튜플 예시

for i in (1,2,3,4,5):
    print(i)
```

```python
# 문자열의 글자 수 세기
# == : 같아?
# != : 같지 않아?

text = "Hello, World!  "
count = 0
for char in text:
    if char != " ":
        count = count + 1
print(count)
```

# 3. while 반복문

## 3.1. while 반복문 예시
- 조건을 제시하고 참인 경우(True)에만 반복을 실행함.


```python
# 1부터 10까지의 합 구하기

n = 1
total = 0
while n <= 10:
    total += n # total = 3
    n += 1 # n = 3
print(total)


# 동일한 결과
sum([i for i in range(1,101)])
```

```python
# 사용자 입력값이 특정 문자열이 아닐 때까지 반복하기
answer = ""
while answer != "no":
    answer = input("Do you want to continue? (yes/no) ")
print('it is done')
```

```python
# 무한 반복문을 사용하여 게임 진행하기

while True:
    user_choice = input("Rock, paper, or scissors? ")

    # 게임 진행 로직
    if user_choice == "quit":
        break
```

```python
# continue, break 키워드 사용하기

while True:
    user_choice = int(input('숫자를 입력하세요\n'))
    if user_choice == 0:
        print(user_choice)
        print('아까는 왜 안됐어??')
        continue
        print('여기서부터는 출력이 안되어야 해!!')
        print('그래 이게 맞지!!')
    else:
        print(user_choice)
        break
```

```python
# 이중 while문 사용하기

n = 0

while n <= 100:
    while n <= 10:
        print(n)
        n += 1
    n += 10
    print(n)
```

# 4. 예외 처리

* 예외 처리란 프로그램이 실행 중에 발생할 수 있는 예외 상황을 처리하는 방법을 말함.
* 예외는 프로그램 실행 중에 발생하는 오류나 예상치 못한 상황을 의미함.
* 이러한 예외 상황에 대처하기 위해 파이썬은 try, except, finally 등의 예외 처리 구문을 제공함.
* 예외 처리를 통해 프로그램 실행 중에 예외가 발생해도 프로그램이 비정상적으로 종료되지 않도록 하거나, 오류에 대한 보다 유용한 정보를 출력하거나, 오류 상황을 처리할 수 있는 기회를 제공할 수 있음.

## 4.1. 예외 처리 예시

```python
# 에러 발생 예시

num = int(input("나누기를 할 정수를 입력하세요: "))
result = 10 / num
print("결과:", result)

result = 10

num2 = int(input("곱하기를 할 정수를 입력하세요."))
result *= num2
print("곱하기 결과:", result)
```

```python
# 예외 처리 기본 구문

try:
    num = int(input("나누기를 할 정수를 입력하세요: "))
    result = 10 / num
    print("결과:", result)
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except ValueError:
    print("올바른 숫자를 입력하세요.")

# except:
#     print('에러가 발생했습니다.')
# except (ZeroDivisionError, ValueError):
#     print('0으로 나누었거나 올바른 value type이 아닙니다.')
```

```python
# 여러 예외를 하나의 'except'블록에서 처리

try:
    file = open("nonexistent.txt", "r")
    content = file.read()
    file.close()
except (FileNotFoundError, PermissionError):
    print("파일을 찾을 수 없거나 접근 권한이 없습니다.")
```

```python
# 'else'와 'finally' 구문 활용

try:
    num = int(input("나눌 수를 입력하세요: "))
    result = 10 / num
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
else:
    print("결과:", result)
finally:
    print("프로그램이 종료되었습니다.")
```

```python
# assert문을 사용한 예외 처리

def divide(x, y):
    assert y != 0, "0으로 나눌 수 없습니다."
    return x / y

try:
    result = divide(10, 0)
except AssertionError as e:
    print(e)
```

```python
# 사용자 정의 예외
# raise는 에러를 발생시키는 기능을 수행함.
# "음수는 입력할 수 없습니다 " 문구는 except NegativeNumberError as e 구문에서 print(e) 됨.
# __str__() 때문에 생기는 문구는 아님. 헷갈리지 말것.

class NegativeNumberError(Exception):
    def __init__(self, value):
        self.value = value
    def __str__(self):
        return f"음수는 입력할 수 없습니다: {self.value}"

try:
    num = int(input("양의 정수를 입력하세요: "))
    if num < 0:
        raise NegativeNumberError(num)
except ValueError:
    print("올바른 숫자를 입력하세요.")
except NegativeNumberError as e:
    print(e)
```
